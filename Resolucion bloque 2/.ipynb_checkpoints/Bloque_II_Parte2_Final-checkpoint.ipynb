# Bloque II — Parte 2 (Entrega Final)

Modelos de regresión, comparación y conclusiones.

```
import warnings
warnings.filterwarnings("ignore")

from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

%matplotlib inline
```

```
DATA_DIR = Path(".")
df = pd.read_csv(DATA_DIR / "ventas_mayo_2026.csv", parse_dates=["fecha"])

df["canal"] = df["canal"].fillna("desconocido")
df["importe"] = df["importe"].fillna(df["importe"].median())
df["dia_semana"] = df["fecha"].dt.dayofweek

df.head()
```

## Definición de variables

```
X = df[["region","canal","unidades","precio_unitario","descuento","dia_semana"]]
y = df["importe"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

## Modelos

```
num_cols = ["unidades","precio_unitario","descuento","dia_semana"]
cat_cols = ["region","canal"]

pre = ColumnTransformer([
    ("num", StandardScaler(), num_cols),
    ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols)
])

lin_model = Pipeline([("pre", pre), ("model", LinearRegression())])

rf_model = Pipeline([
    ("pre", pre),
    ("model", RandomForestRegressor(n_estimators=300, max_depth=10, random_state=42))
])
```

```
lin_model.fit(X_train, y_train)
pred_lin = lin_model.predict(X_test)

rf_model.fit(X_train, y_train)
pred_rf = rf_model.predict(X_test)
```

```
metrics = pd.DataFrame({
    "modelo": ["LinearRegression","RandomForest"],
    "MAE": [mean_absolute_error(y_test, pred_lin), mean_absolute_error(y_test, pred_rf)],
    "RMSE": [mean_squared_error(y_test, pred_lin, squared=False), mean_squared_error(y_test, pred_rf, squared=False)],
    "R2": [r2_score(y_test, pred_lin), r2_score(y_test, pred_rf)]
})

metrics.round(3)
```

```
errores = pd.DataFrame({"real": y_test, "predicho": pred_rf})
errores["error"] = errores["real"] - errores["predicho"]

errores.sort_values("error", key=abs, ascending=False).head(10)
```

```
plt.figure()
plt.scatter(errores["real"], errores["predicho"], alpha=0.6)
plt.plot([errores["real"].min(), errores["real"].max()],
         [errores["real"].min(), errores["real"].max()])
plt.xlabel("Real")
plt.ylabel("Predicho")
plt.title("Real vs Predicho")
plt.show()
```

## Conclusiones

El modelo Random Forest presenta mejor rendimiento en términos de RMSE y R², lo que indica una mayor precisión en la predicción del importe.

La regresión lineal, aunque menos precisa, ofrece mayor interpretabilidad, permitiendo entender la relación entre variables.

El ajuste de hiperparámetros en Random Forest (max_depth=10) ayuda a mejorar la generalización y evitar sobreajuste.

Desde un punto de vista de negocio, este modelo permite estimar el valor de futuras ventas con suficiente precisión para apoyar la toma de decisiones comerciales.

No siempre debe elegirse el modelo con menor error, ya que factores como interpretabilidad y facilidad de uso también son relevantes.
